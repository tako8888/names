```python



year_birth = birth_df.groupby('Year').sum().drop(columns = 'State')


def calculate_share(row):
    target_year = row['Year']
    total_birth = year_birth.loc[target_year]
    answer = row['Births'] / total_birth
    return answer

total_df['share_in_year'] = total_df.apply(calculate_share, axis=1).copy()

    
# total_df['share_in_year'] = 
```

```python
top10_df = (total_df.groupby(['Year','Sex'], group_keys=False)
     .apply(lambda x: x.nlargest(10, ['Births']))
     .reset_index()
     .drop(columns='index'))
top10_df
```
